```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import stemgraphic # for stem-and-leaf plot drawing
import scipy.stats as stats

#x = np.array(list(map(float, input("Enter Numbers: ").split())))
#y = np.array(list(map(float, input("Enter Numbers: ").split())))
x = np.array( [2, 6, 8, 12, -19, 30, 0, -5, 7, 16, 23, 38, -29, 35, 1, -28])
y = np.array([9, 2, -4, 42, 9, 23, -3, -6, 5, 22, -14, 51, 65, 3, -16, -3])


print("Data set of x =",x)
print("Data set of y =",y)
print("\nMean:\n")
print(f"Mean of Data set x is {np.mean(x):.4f}.")
print(f"Mean of Data set y is {np.mean(y):.4f}.\n")
print("Median:\n")
print (f"Median of Data set x is {np.median(x):.1f}.")
print (f"Median of Data set y is {np.median(y):.1f}.")

#--------------------------------Seperated Histrogram Drawing----------------------------------------#
print("\nSeperated Histrogram Drawing:\n")

#------------------------- Create Figure for x Data Set---------------------------------#

plt.figure(figsize=(6,4), dpi=300) # High resolution (300 DPI for articles)
# plot Histrogram
plt.hist(x, bins=10, color='skyblue', edgecolor='black')
# Add mean line
plt.axvline(np.mean(x), color="red", linestyle="dashed", linewidth=1.5, label= f"Mean = {np.mean(x):.2f}")


# Labels and title
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Histogram of data set x", fontsize=14, weight='bold')


# Legand 
plt.legend()

# Ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Tight layout for clean look
plt.tight_layout()

# Save figure (high quality PNG for article)
plt.savefig("Histogram of x.png", dpi=300)

# show plot
plt.show()

#------------------------------------------ Create Figure for y Data Set ------------------------------#

plt.figure(figsize=(6,4), dpi=300) # High resolution (300 DPI for articles)
# plot Histrogram
plt.hist(y, bins=10, color='skyblue', edgecolor='black')
# Add mean line
plt.axvline(np.mean(y), color="red", linestyle="dashed", linewidth=1.5, label= f"Mean = {np.mean(y):.2f}")


# Labels and title
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Histogram of data set y", fontsize=14, weight='bold')


# Legand 
plt.legend()

# Ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Tight layout for clean look
plt.tight_layout()

# Save figure (high quality PNG for article)
plt.savefig("Histogram of y.png", dpi=300)

# show plot
plt.show()

#--------------------------------Seperated Histrogram Drawing----------------------------------------#
print("\nSide by side Histrogram Drawing:\n")
#--------------------------------Side by side Histrogram Drawing----------------------------------------#

plt.figure(figsize=(6,4), dpi=300) # High resolution (300 DPI for articles)

# Dataset x
plt.subplot(1,2,1)
plt.hist(x, bins='auto', color='steelblue', edgecolor='black', alpha=0.8)
plt.axvline(np.mean(x), color='red', linestyle='dashed', linewidth=1.5, label=f"Mean = {np.mean(x):.2f}")
plt.title("Histrogram of Data set x")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# Dataset y
plt.subplot(1,2,2)
plt.hist(data2, bins='auto', color='orange', edgecolor='black', alpha=0.8)
plt.axvline(np.mean(y), color='red', linestyle='dashed', linewidth=1.5, label=f"Mean = {np.mean(y):.2f}")
plt.title("Histrogram of Data set y")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.savefig("xy_hist_side_by_side.png", dpi=300)
plt.show()

#--------------------------------Side by side Histrogram Drawing----------------------------------------# 

#--------------------------------Create stem-and-leaf plot for x dataset----------------------------------------------#
#import stemgraphic

#Create figure
fig, ax = stemgraphic.stem_graphic(x, scale=20)  # scale controls figure size

# Title & formatting
plt.title("Stem-and-Leaf Plot of Dataset x", fontsize=14, weight="bold")

# Save high-resolution figure
plt.savefig("stem_leaf_plot_of_Dataset_x.png", dpi=300, bbox_inches="tight")  # PNG
plt.savefig("stem_leaf_plot_of_Dataset_x.tif", dpi=300, bbox_inches="tight")  # TIFF (for journals)

plt.show()


#--------------------------------Create stem-and-leaf plot for x dataset----------------------------------------------#


#--------------------------------Create stem-and-leaf plot for y dataset----------------------------------------------#

#import stemgraphic

#Create figure
fig, ax = stemgraphic.stem_graphic(y, scale=20)  # scale controls figure size

# Title & formatting
plt.title("Stem-and-Leaf Plot of Dataset y", fontsize=14, weight="bold")

# Save high-resolution figure
plt.savefig("stem_leaf_plot_of_Dataset_y.png", dpi=300, bbox_inches="tight")  # PNG
plt.savefig("stem_leaf_plot_of_Dataset_y.tif", dpi=300, bbox_inches="tight")  # TIFF (for journals)

plt.show()


#--------------------------------Create back to back stem-and-leaf plot for xy dataset----------------------------------------------#

def back_to_back_stem_leaf_plot(x, y, title="Back-to-Back Stem-and-Leaf Plot"):
    # Sort datasets
    Sorted_Data_x = sorted(x)
    Sorted_Data_y = sorted(y)

    # Create stems and leaves
    def make_stem_leaf(data):
        stems = {}
        for num in data:
            stem, leaf = divmod(int(num), 10)
            stems.setdefault(stem, []).append(leaf)
        return stems

    stems1 = make_stem_leaf(Sorted_Data_x)
    stems2 = make_stem_leaf( Sorted_Data_y)

    all_stems = sorted(set(stems1.keys()) | set(stems2.keys()))

    # Prepare data for plotting
    left_counts = [-len(stems1.get(stem, [])) for stem in all_stems]  # negative for left side
    right_counts = [len(stems2.get(stem, [])) for stem in all_stems]

    # Plot
    fig, ax = plt.subplots(figsize=(6, 6))

    ax.barh(all_stems, left_counts, color="skyblue", label="Dataset x")
    ax.barh(all_stems, right_counts, color="salmon", label="Dataset y")

    # Formatting
    ax.set_title(title, fontsize=14, weight="bold")
    ax.set_xlabel("leaf", fontsize=12)
    ax.set_ylabel("Stem (Tens)", fontsize=12)
    ax.legend()
    ax.grid(axis="x", linestyle="--", alpha=0.6)

    # Tidy up axes
    ax.set_yticks(all_stems)
    ax.set_xticks(np.arange(min(left_counts)-1, max(right_counts)+2, 1))
    ax.axvline(0, color="black", linewidth=1.2)

    plt.tight_layout()
    
    # Save high-resolution figure
    plt.savefig("back_to_back_stem_leaf.png", dpi=300, bbox_inches="tight")
    
    plt.show()

back_to_back_stem_leaf_plot(data1, data2, title="Back-to-Back Stem-and-Leaf Plot")

#--------------------------------Create back to back stem-and-leaf plot for xy dataset----------------------------------------------#

#--------------------------------Ordered Dataset------------------------------------------------------------------------------------#

Ordered_Data_x = np.sort(x)
Ordered_Data_x = np.sort(y)

print("\nData set of x:", x)
print("Therefore, ordered data set of x:", x)

print("\nData set of y:", y)
print("Therefore, ordered data set of y:", y)

#--------------------------------Ordered Dataset------------------------------------------------------------------------------------#

#----------------------------percentile & Quantile calculation of x & y Data set---------------------------------------------------------------#

# Define percentiles: deciles (0%, 10%, …, 100%)
percentiles = np.arange(0, 1.1, 0.1)

# Compute quantiles
quantile_values_x = np.quantile(x, percentiles)
quantile_values_y = np.quantile(y, percentiles)

# Format results into a table
results = pd.DataFrame({
    "Percentile": [f"{int(p*100)}%" for p in percentiles],
    "Value of x": [f"{q:.2f}" for q in quantile_values_x],
    "Value of y": [f"{q:.2f}" for q in quantile_values_y]
})

print("\n", results.to_string(index=False),"\n")

#----------------------------percentile & Quantile calculation of x & y Data set---------------------------------------------------------------#

#----------------------------Quantile calculation of x & y Data set for given percentile---------------------------------------------------------------#

# Example dataset
data = np.array([12, 14, 15, 15, 16, 18, 21, 22, 23, 25, 27, 29])

# Given percentile positions (in fraction form, 0–1)
p = [0.1, 0.34, 0.68, 0.93]

# Compute quantile values at those percentiles from x dataset
Quantile_values_x = np.quantile(x, p)
# Compute quantile values at those percentiles from y dataset
Quantile_values_y = np.quantile(y, p)

# Build a table
results = pd.DataFrame({
    "Percentile": [f"{int(prob*100)}th" for prob in p],
    "x_Value": [f"{q:.2f}" for q in Quantile_values_x],
    "y_Value": [f"{q:.2f}" for q in Quantile_values_y],
})

print("\n", results.to_string(index=False),"\n")

#----------------------------Quantile calculation of x & y Data set for given percentile---------------------------------------------------------------#

#----------------------------Interquantile calculation of x & y Data set for given percentile---------------------------------------------------------------#

# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1_x = np.percentile(x, 25)
Q3_x = np.percentile(x, 75)
Q1_y = np.percentile(y, 25)
Q3_y = np.percentile(y, 75)

# Interquartile Range
IQR_x = Q3_x - Q1_x
IQR_y = Q3_y - Q1_y

print(f"Q1 of x (25th percentile): {Q1_x:.2f}")
print(f"Q3 of x (75th percentile): {Q3_x:.2f}\n")
print(f"Q1 of y (25th percentile): {Q1_y:.2f}")
print(f"Q3 of y (75th percentile): {Q3_y:.2f}\n")
print(f"IQR of x: {IQR_x:.2f}\n")
print(f"IQR of y: {IQR_y:.2f}\n")

#----------------------------Interquantile calculation of x & y Dataset for given percentile---------------------------------------------------------------#

#----------------------------Statistics calculation of x & y Dataset---------------------------------------------------------------#

# Function to compute statistics for any dataset. In Example: x & Y of above Data set.
def compute_stats(data):
    mean = np.mean(data)
    median = np.median(data)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    variance = np.var(data, ddof=1)  # sample variance
    std_dev = np.std(data, ddof=1)   # sample SD
    MAD = np.mean(np.abs(data - mean))
    return [mean, median, Q1, Q3, IQR, variance, std_dev, MAD]

# Compute stats for x & y datasets
stats1 = compute_stats(x)
stats2 = compute_stats(y)

# Create table
stats_table = pd.DataFrame({
    "Statistic": ["Mean", "Median", "Q1 (25th percentile)", "Q3 (75th percentile)", "IQR", "Variance (sample)", "Std Dev (sample)","MAD"],
    "Dataset x": [f"{n:.2f}" for n in stats1],
    "Dataset y": [f"{n:.2f}" for n in stats2]
})

# Display table
print(stats_table.to_string(index=False), "\n")

#----------------------------Boxplot calculation & plot of x & y Dataset---------------------------------------------------------------#

datasets = {"Data_set x": x, "Data_set y": y}
summary = {}

# Calculate boxplot stats for each dataset
for name, data in datasets.items():
    mean = np.mean(data)
    median = np.median(data)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    variance = np.var(data, ddof=1)  # sample variance
    std_dev = np.std(data, ddof=1)   # sample SD
    MAD = np.mean(np.abs(data - mean))
    
    # Whiskers (using Tukey’s rule: 1.5 * IQR)
    lower_whisker = np.min(data[data >= Q1 - 1.5*IQR])
    upper_whisker = np.max(data[data <= Q3 + 1.5*IQR])
    
    # Outliers
    outliers = data[(data < lower_whisker) | (data > upper_whisker)]
    
    summary[name] = {
        "Mean": mean,
        "Q1 (25%)": Q1,
        "Median (50%)": median,
        "Q3 (75%)": Q3,
        "IQR": IQR,
        "Variance": variance,
        "Std. Deviation": std_dev,
        "MAD": MAD,
        "Lower Whisker": lower_whisker,
        "Upper Whisker": upper_whisker,
        "Outliers": list(outliers)
    }

# Convert to pandas DataFrame
summary_df = pd.DataFrame(summary).T
print(summary_df)

# Draw Boxplot
plt.figure(figsize=(6,4), dpi=300) # High resolution (300 DPI for articles)

#Formatting of Boxplot
plt.boxplot(datasets.values(), vert=True, patch_artist=True, tick_labels=datasets.keys(),)
plt.title("Boxplot Comparison of Datasets",fontsize=14, weight="bold")
plt.xticks(fontsize=12)
plt.yticks(fontsize=10)
plt.ylabel("Values", fontsize=12)


plt.tight_layout()
plt.savefig("xy_Boxplot_side_by_side.png", dpi=300)
plt.show()

#----------------------------Boxplot calculation & plot of x & y Dataset---------------------------------------------------------------#

# Draw Boxplot
plt.figure(figsize=(6,4), dpi=300) # High resolution (300 DPI for articles)

plt.scatter(x, y, color="green", alpha=0.7, edgecolor="black")
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("X values", fontsize=12, weight="bold")
plt.ylabel("Y values", fontsize=12, weight="bold")
plt.title("Scatterplot with Correlation", fontsize=14, weight="bold")

# Compute correlation
corr = np.corrcoef(x, y)[0, 1]

# Annotate correlation on the plot
plt.text(max(x-10), max(y), f"r = {corr:.2f}", fontsize=12, weight="bold", color="red")

plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.savefig("Scatterplot with Correlation.png", dpi=300)
plt.show()

#----------------------------Quantile calculation & plot of x & y Dataset---------------------------------------------------------------#


# Step 1: number of data points
n = len(x)

# Step 2: sort the dataset
s = np.sort(x)

# Step 3: compute i (1, 2, ..., n)
i = np.arange(1, n+1)

# Step 4: compute plotting positions u = (i - 0.5) / n
u = (i - 0.5) / n

# Draw Boxplot
plt.figure(figsize=(6,4), dpi=300) # High resolution (300 DPI for articles)

# Step 5: plot quantile plot
plt.plot(u, s, marker="o", linestyle="--", color="blue")

plt.xlabel("Quantiles (u)", fontsize=12, weight="bold")
plt.ylabel("Sorted Data (s)", fontsize=12, weight="bold")
plt.title("Quantile Plot", fontsize=14, weight="bold")
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout()
plt.savefig("quantile plot.png", dpi=300)
plt.show()

#----------------------------Quantile calculation & plot of x & y Dataset---------------------------------------------------------------#

#----------------------------Q-Q calculation & plot of x & y Dataset---------------------------------------------------------------#

# Compute theoretical quantiles from Normal distribution
# Generate probabilities (same as plotting positions)
theoretical_q = stats.norm.ppf(u, loc=np.mean(s), scale=np.std(s))

# Draw Boxplot
plt.figure(figsize=(6,4), dpi=300) # High resolution (300 DPI for articles)

# Plot Q-Q plot
plt.scatter(theoretical_q, s, color="blue", edgecolor="black", label="Data Quantiles")
plt.plot([min(theoretical_q), max(theoretical_q)],
         [min(s), max(s)],
         color="red", linestyle="--", label="y = x (Perfect Fit)")

plt.xlabel("Theoretical Quantiles (Normal)", fontsize=12, weight="bold")
plt.ylabel("Sample Quantiles (Data)", fontsize=12, weight="bold")
plt.title("Q–Q Plot vs Normal Distribution", fontsize=14, weight="bold")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)


plt.tight_layout()
plt.savefig("Q-Q plot.png", dpi=300)
plt.show()

#----------------------------Q-Q calculation & plot of x & y Dataset---------------------------------------------------------------#
```
